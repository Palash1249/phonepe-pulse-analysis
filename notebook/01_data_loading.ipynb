```python
import os
import json
import pandas as pd

print("All imports successful")
```

```python
records = []
base = "../pulse/data/aggregated/transaction/country/india/state/"

for state in os.listdir(base):
    state_path = os.path.join(base, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for file in os.listdir(year_path):
            quarter = file.replace(".json", "")
            file_path = os.path.join(year_path, file)
            with open(file_path) as f:
                content = json.load(f)
                for item in content["data"]["transactionData"]:
                    records.append({
                        "state": state,
                        "year": int(year),
                        "quarter": int(quarter),
                        "transaction_type": item["name"],
                        "count": item["paymentInstruments"][0]["count"],
                        "amount": item["paymentInstruments"][0]["amount"]
                    })

df_transactions = pd.DataFrame(records)
print(f"Rows loaded: {len(df_transactions)}")
print(df_transactions.head())
```

```python
user_records = []
base_user = "../pulse/data/aggregated/user/country/india/state/"

for state in os.listdir(base_user):
    state_path = os.path.join(base_user, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for file in os.listdir(year_path):
            quarter = file.replace(".json", "")
            file_path = os.path.join(year_path, file)
            with open(file_path) as f:
                content = json.load(f)
                data = content["data"]
                user_records.append({
                    "state": state,
                    "year": int(year),
                    "quarter": int(quarter),
                    "registered_users": data["aggregated"]["registeredUsers"],
                    "app_opens": data["aggregated"]["appOpens"]
                })

df_users = pd.DataFrame(user_records)
print(f"Rows loaded: {len(df_users)}")
print(df_users.head())
```

```python
print("=== TRANSACTIONS ===")
print(df_transactions.info())
print(df_transactions.isnull().sum())
print(df_transactions["transaction_type"].unique())
print(df_transactions["year"].unique())

print("\n=== USERS ===")
print(df_users.info())
print(df_users.isnull().sum())
```

```python
df_transactions.to_csv("../data/agg_transactions.csv", index=False)
df_users.to_csv("../data/agg_users.csv", index=False)
print("Saved to /data/ folder")
```
